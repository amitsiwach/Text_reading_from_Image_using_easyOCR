```python
import easyocr
```

```python
reader = easyocr.Reader(['en', 'hi'])
```

```python
Image("/content/drive/MyDrive/ColabNotebooks/Image.jpg")
```

```python
output = reader.readtext("/content/drive/MyDrive/ColabNotebooks/Image.jpg") 
```

```python
output
```

```python
cord = output[-1][0]
cord
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
```

```python
x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
```

```python
image = cv2.imread("/content/drive/MyDrive/ColabNotebooks/Image.jpg")
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
